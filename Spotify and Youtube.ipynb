```python
# Eksik değerlere sahip olan sütunlardaki az sayıdaki eksik değerleri doldurma 
columns_with_few_missing = ['Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo', 'Duration_ms']
for column in columns_with_few_missing:
    data[column].fillna(data[column].mean(), inplace=True)
      
#'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo' ve 'Duration_ms' gibi sütunlarda 
#sadece 2 eksik değer vardır. Bu kadar az sayıda eksik değer olduğu için, eksik değerlerin sırasıyla ilgili sütunun 
#ortalaması veya medyanı ile doldurulması mantıklı bir yaklaşımdır. 
#Bu yöntem, verinin genel dağılımını korumaya ve bilgi kaybını minimize etmeye yardımcı olur. 
#Bu şekilde, eksik değerlerin yerine geçecek değerlerin belirlenmesi, veri setinin analiz veya modelleme için 
#daha uygun hale getirilmesine katkıda bulunur.
```

```python
# Birçok eksik değeri olan sütunların listesini göstermektedir
columns_with_many_missing = ['Channel', 'Views', 'Likes', 'Comments', 'Licensed', 'official_video', 'Stream']

# Belirtilen sütunlardaki eksik değerlere sahip olan satırları veri kümesinden çıkarma
for column in columns_with_many_missing:
    data.dropna(subset=[column], inplace=True)
    
#'Channel', 'Views', 'Likes', 'Comments', 'Licensed', 'official_video' ve 'Stream' sütunlarında eksik değerlerin bulunması 
#durumunda, bu eksik değerlere sahip olan satırların silinmesi uygun bir seçenek olabilir. 
#Ancak, bu yaklaşım veri setinin küçük bir kısmını temsil eden eksik değerleri kaldırdığı ve analizi önemli ölçüde 
#etkilemediği durumlarda uygun olmaktadır.
#burada da bunu yapmanın uygun olacağı sonucuna varıp yukarıda ifade edilen sütunlarda eksik değerlere sahip olan satırlar 
#veri setinden çıkarılarak, eksik değerlerden arındırılmış bir veri seti elde edilmiştir. 
#Bu işlem, eksik değerlerin analiz veya modelleme süreçlerini etkilememesini sağlar ve veri setinin doğruluğunu artırır.     
```

```python
data_grouped = data.groupby('Artist')[['Views', 'Stream']].sum().reset_index()
```

```python
data_grouped = data_grouped.sort_values(by='Views', ascending=False)[:10]
```

```python
data_melted = pd.melt(data_grouped, id_vars=['Artist'], var_name='Platform', value_name='Counts')
```

```python
sns.barplot(y='Artist', x='Counts', hue='Platform', data=data_melted, orient = 'h')
plt.xticks(rotation=0)
plt.xlabel('Artist')
plt.ylabel('Counts')
plt.title('Top 10 Artists on YouTube and Spotify');
```
